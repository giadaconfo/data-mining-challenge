```python
import numpy as np
import pandas as pd
train_data = pd.read_csv('./data/train.csv')
```

```python
train_data['Date']=pd.to_datetime(train_data['Date'], format='%d/%m/%Y')
train_data['Day'] = pd.DatetimeIndex(train_data['Date']).day
train_data['Month'] = pd.DatetimeIndex(train_data['Date']).month
train_data['Year'] = pd.DatetimeIndex(train_data['Date']).year
train_data=train_data.drop('Date',axis=1)
```

```python
train_data.head()
```

```python
result = pd.DataFrame(train_data['StoreID'])
result['Day']=train_data['Day']
result['NumberOfSales'] = train_data['NumberOfSales']
#Group by Month
result =result.groupby(['Day'], as_index=False)['NumberOfSales'].sum()
```
