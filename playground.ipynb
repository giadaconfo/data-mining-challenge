```python
# dataframe management
import pandas as pd             

# numerical computation
import numpy as np

# visualization library
import seaborn as sns

# plot 
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
df = pd.read_csv('data/train.csv')

df.head(5)
```

```python
print(df.dtypes)
numerical_var = df.select_dtypes(exclude=[object])
categorical_var = df.select_dtypes(include=[object])
```

```python
numerical_var.describe()
```

```python

categorical_variables_mask = df.dtypes==object
categorical_variables = df.columns[categorical_variables_mask]
numerical_variables = df.columns[~categorical_variables_mask]
print("%d Categorical Variables\n%d Numerical Variables\n"%(len(categorical_variables),len(numerical_variables)))
```

```python

categorical_variables_mask = df.dtypes==object
categorical_variables = df.columns[categorical_variables_mask]
# date doesnt provide any relevant ifo 
categorical_variables = categorical_variables.drop('Date')
categorical_variables.shape
for cat in categorical_variables:
    print(df[cat].unique())
    print("\n")
```

```python
cov=df[numerical_variables].corr(method='pearson')
plt.figure(figsize = (16,16))
sns.heatmap(cov,square=True,annot=True,cmap="Blues",cbar=False,xticklabels=True, yticklabels=True)
```

```python
cm = sns.clustermap(df[numerical_variables], annot=False, center=0, cmap="Blues", figsize=(8, 8))
cm.cax.set_visible(False)
```

```python
plt.figure(figsize=(10,8))
for v in numerical_variables:
    plt.plot(df[v])
plt.legend(loc='best')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df[numerical_variables])
df_normalized_values = scaler.transform(df[numerical_variables])
df_normalized = pd.DataFrame(data = df_normalized_values, columns=numerical_variables)
```
